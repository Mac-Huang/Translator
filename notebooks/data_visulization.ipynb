```python
# Import necessary libraries
import os
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
import matplotlib.font_manager as fm

# Set up Matplotlib to support Chinese characters
plt.rcParams['font.sans-serif'] = ['SimHei']  # Use 'SimHei' for Chinese support on most systems
plt.rcParams['axes.unicode_minus'] = False    # Ensure minus signs display correctly

# Set the path to your data directory
data_dir = '../data/MSLT_Corpus/Data/MSLT_Dev_ZH'

# Load English T2 and Chinese T3 files
english_texts_t3 = []
chinese_texts_t2 = []

# Load T2 English files
for filename in os.listdir(data_dir):
    if filename.endswith(".T2.ch.snt"):
        with open(os.path.join(data_dir, filename), 'r', encoding='utf-16') as f:
            chinese_texts_t2.extend(f.readlines())

# Load T3 Chinese files
for filename in os.listdir(data_dir):
    if filename.endswith(".T3.en.snt"):
        with open(os.path.join(data_dir, filename), 'r', encoding='utf-16') as f:
            english_texts_t3.extend(f.readlines())


# Display a sample of sentences from English T2 and Chinese T3 in a table
sample_size = 20  # Number of sentences to display
df = pd.DataFrame({
    "Chinese T2": chinese_texts_t2[:sample_size],
    "English T3": english_texts_t3[:sample_size]
})

# Display the DataFrame
df_display = df.style.set_properties(**{'text-align': 'left'})
df_display.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
df_display
```

```python
# 1. Sentence Length Distribution (Token Count)
# Calculate sentence lengths (in tokens)


english_lengths_t2 = [len(sentence.split()) for sentence in english_texts_t2]
chinese_lengths_t3 = [len(sentence.split()) for sentence in chinese_texts_t3]

# Plot distribution
plt.figure(figsize=(12, 6))
plt.hist(english_lengths_t2, bins=30, alpha=0.5, label="English T2")
plt.hist(chinese_lengths_t3, bins=30, alpha=0.5, label="Chinese T3")
plt.xlabel("Sentence Length (Tokens)")
plt.ylabel("Frequency")
plt.title("Sentence Length Distribution (English T2 vs Chinese T3)")
plt.legend()
plt.show()
```

```python
# 2. Top Token Frequency
# Tokenize and count frequencies
english_tokens_t2 = Counter(" ".join(english_texts_t2).split())
chinese_tokens_t3 = Counter(" ".join(chinese_texts_t3).split())

# Get the most common tokens
common_english_t2 = english_tokens_t2.most_common(20)
common_chinese_t3 = chinese_tokens_t3.most_common(20)

# Separate words and counts for plotting
english_tokens, english_counts = zip(*common_english_t2)
chinese_tokens, chinese_counts = zip(*common_chinese_t3)

# Plot token frequency for English T2
plt.figure(figsize=(12, 6))
plt.bar(english_tokens, english_counts, color='blue', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel("Token")
plt.ylabel("Frequency")
plt.title("Top 20 Common Tokens in English T2")
plt.show()

# Plot token frequency for Chinese T3
plt.figure(figsize=(12, 6))
plt.bar(chinese_tokens, chinese_counts, color='green', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel("Token")
plt.ylabel("Frequency")
plt.title("Top 20 Common Tokens in Chinese T3")
plt.show()
```

```python
# 3. Unique Tokens per Sentence
# Count unique tokens per sentence
unique_english_tokens_t2 = [len(set(sentence.split())) for sentence in english_texts_t2]
unique_chinese_tokens_t3 = [len(set(sentence.split())) for sentence in chinese_texts_t3]

# Plot unique tokens per sentence
plt.figure(figsize=(12, 6))
plt.hist(unique_english_tokens_t2, bins=30, alpha=0.5, label="English T2 Unique Tokens")
plt.hist(unique_chinese_tokens_t3, bins=30, alpha=0.5, label="Chinese T3 Unique Tokens")
plt.xlabel("Number of Unique Tokens per Sentence")
plt.ylabel("Frequency")
plt.title("Distribution of Unique Tokens per Sentence (English T2 vs Chinese T3)")
plt.legend()
plt.show()
```

```python
# 4. Character Count Distribution (for Chinese)
# Calculate character counts
english_char_count_t2 = [len(sentence) for sentence in english_texts_t2]
chinese_char_count_t3 = [len(sentence) for sentence in chinese_texts_t3]

# Plot character count distribution
plt.figure(figsize=(12, 6))
plt.hist(english_char_count_t2, bins=30, alpha=0.5, label="English T2 Character Count")
plt.hist(chinese_char_count_t3, bins=30, alpha=0.5, label="Chinese T3 Character Count")
plt.xlabel("Character Count")
plt.ylabel("Frequency")
plt.title("Character Count Distribution (English T2 vs Chinese T3)")
plt.legend()
plt.show()
```
